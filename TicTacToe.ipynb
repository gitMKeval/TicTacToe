```python
from random import randrange

print("Welcome User in TIC TAC TOE \n") 

def display_board(board):
    
    print("+-----------+-----------+-----------+\n")
    print("|    ",board[0][0],"    +    ",board[0][1],"    +    ",board[0][2],"    |\n")
    print("+-----------+-----------+-----------+\n")
    print("|    ",board[1][0],"    +    ",board[1][1],"    +    ",board[1][2],"    |\n")
    print("+-----------+-----------+-----------+\n")
    print("|    ",board[2][0],"    +    ",board[2][1],"    +    ",board[2][2],"    |\n")
    print("+-----------+-----------+-----------+\n")

def free_field(board):
    free=[]
    for row in range(3):
        for col in range(3):
            if board[row][col] not in ["X","O"]:
                free.append((row,col))
    return free         

def enter_move(board):
    ok = False 
    while not ok:
        move = int(input("Enter your move : "))
        ok = move>=1 and move <=9
        if not ok:
            print(" Incorrect Input ")
            continue
        move = move -1
        row=move//3
        col=move%3
        sign=board[row][col]
        ok = sign not in ["O","X"]
        if not ok:
            print("Field already occupied choose another move : ")
            continue
    board[row][col] = "O"
    
def victor(board,sign):
    if sign == "X" :
        win = "Program"
    elif sign == "O" :
        win = "User"
    else:
        win = None
    cross1=cross2=True
    for rc in range(3):
        if board[rc][0]==sign and board[rc][1]==sign and board[rc][2]==sign :
            return win
        if board[0][rc]==sign and board[1][rc]==sign and board[2][rc]==sign :
            return win
        if board[rc][rc] != sign:
            cross1 = False
        if board[rc][2-rc] != sign:
            cross2 = False
    if cross1 or cross2 :
        return win
    return None

def program_move(board):
    free=free_field(board)
    count=len(free)
    if count > 0 :
        p_move = randrange(count)
        row,col=free[p_move]
        board[row][col] = "X"

    
board = [[3*j+i+1 for i in range(3)] for j in range(3)]
board[1][1] = "X"

free=free_field(board)
u_move = True

while len(free):
    display_board(board)
    if u_move :
        enter_move(board)
        winner = victor(board,"O")
    else:
        program_move(board)
        winner = victor(board,"X")
    if winner != None :
        break
    u_move = not u_move
    free=free_field(board)

display_board(board)
if winner == "Program" :
    print("Python Won !")
elif winner == "User" :
    print("User Won !")
else :
    print("Tie")
```
